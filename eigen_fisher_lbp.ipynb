```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2.face
```

```python
faces = []
responses = []

count = 0

directory = "../dataset/smaller_set_faces_64x128"
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"): 
        thePath = os.path.join(directory, filename)
        print(thePath)
        img = cv2.imread(thePath, 0)
        #faces.append(img.flatten())
        faces.append(img[33:128,0:64])
        
        responses.append(1)
        
        
        #plt.imshow(img[33:128,0:64])
        #plt.show()
        #break
        
        count = count + 1
        
        continue
    else:
        continue
#return
directory = "../dataset/nonfaces"
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"): 
        thePath = os.path.join(directory, filename)
        print(thePath)
        img = cv2.imread(thePath, 0)
        faces.append(img[33:128,0:64])
        #faces.append(img.flatten())
        responses.append(0)
        
        count = count + 1
        
        continue
    else:
        continue


        
print (responses)
```

```python
face_detector =  cv2.face.LBPHFaceRecognizer_create()#cv2.face.EigenFaceRecognizer_create()#cv2.face.FisherFaceRecognizer_create()#cv2.face.EigenFaceRecognizer_create()#cv2.face.LBPHFaceRecognizer_create()
face_detector.train(faces, np.array(responses))

face_detector.save("lbphf_classifier.xml")
```

```python
width = 64
height =95#128

i=0
searchMe =  cv2.imread("../searchMe2_resized.jpg",0)
outputPic = searchMe.copy()

for x in range (0, searchMe.shape[1]- width, 30 ):
    for y in range(0, searchMe.shape[0] -height, 30 ):
        img = searchMe[y:y+height,x:x+width ]
        
        #output = face_detector.predict(img.flatten())
        output = face_detector.predict(img)
        if output[0] == 1:
            outputPic = cv2.rectangle(outputPic, (x, y), (x+width, y+height), (0,255,0), 1)

plt.imshow (outputPic)
plt.show()
cv2.imwrite ("the results.jpg", outputPic)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2.face
```

```python
faces = []
responses = []

count = 0

directory = "../00000-20201121T135044Z-001/00000/cropped_resized"
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"): 
        thePath = os.path.join(directory, filename)
        print(thePath)
        img = cv2.imread(thePath, 0)
        #faces.append(img.flatten())
        faces.append(img)
        
        responses.append(1)
        
        
        #plt.show()
        #break
        
        count = count + 1
        
        continue
    else:
        continue

directory = "../negative_samples/synthesized_2"
for filename in os.listdir(directory):
    if filename.endswith(".jpg") or filename.endswith(".png"): 
        thePath = os.path.join(directory, filename)
        print(thePath)
        img = cv2.imread(thePath, 0)
        faces.append(img)
        #faces.append(img.flatten())
        responses.append(0)
        
        count = count + 1
        
        continue
    else:
        continue


        
print (responses)
```

```python
face_detector =  cv2.face.LBPHFaceRecognizer_create()#cv2.face.EigenFaceRecognizer_create()#cv2.face.FisherFaceRecognizer_create()#cv2.face.EigenFaceRecognizer_create()#cv2.face.LBPHFaceRecognizer_create()
face_detector.train(faces, np.array(responses))
```

```python
for i in range (count):
    print (face_detector.predict(faces[i]))
```

```python
width = 64
height = 128

i=0
searchMe =  cv2.imread("../searchMe_resized.jpg",0)
outputPic = searchMe.copy()

for x in range (0, searchMe.shape[1]- width, 10 ):
    for y in range(0, searchMe.shape[0] -height, 10 ):
        img = searchMe[y:y+height,x:x+width ]
        
        #output = face_detector.predict(img.flatten())
        output = face_detector.predict(img)
        if output[0] == 1:
            outputPic = cv2.rectangle(outputPic, (x, y), (x+width, y+height), (0,255,0), 1)

plt.imshow (outputPic)
plt.show()
cv2.imwrite ("test result girls.jpg", outputPic)
```
